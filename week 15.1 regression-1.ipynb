**Q1. Explain the difference between simple linear regression and multiple linear regression. Provide an
example of each.**

**ANSWER:------**

### Simple Linear Regression

**Definition**: Simple linear regression is a statistical method used to model the relationship between two variables by fitting a linear equation to observed data. One variable is considered an independent variable (predictor or explanatory variable), and the other is a dependent variable (response or outcome variable).

**Equation**: 
\[ y = \beta_0 + \beta_1x + \epsilon \]
Where:
- \( y \) is the dependent variable.
- \( x \) is the independent variable.
- \( \beta_0 \) is the intercept.
- \( \beta_1 \) is the slope.
- \( \epsilon \) is the error term.

**Example**: Predicting a student's exam score based on the number of hours studied.
- **Independent variable (x)**: Hours studied
- **Dependent variable (y)**: Exam score

### Multiple Linear Regression

**Definition**: Multiple linear regression is an extension of simple linear regression that models the relationship between a dependent variable and two or more independent variables.

**Equation**: 
\[ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \ldots + \beta_nx_n + \epsilon \]
Where:
- \( y \) is the dependent variable.
- \( x_1, x_2, \ldots, x_n \) are the independent variables.
- \( \beta_0 \) is the intercept.
- \( \beta_1, \beta_2, \ldots, \beta_n \) are the coefficients.
- \( \epsilon \) is the error term.

**Example**: Predicting a house price based on its size, number of bedrooms, and age.
- **Independent variables (x_1, x_2, x_3)**: Size of the house (square feet), number of bedrooms, age of the house
- **Dependent variable (y)**: House price

### Summary

- **Simple Linear Regression**: One independent variable, one dependent variable.
  - **Example**: Hours studied (independent) predicting exam score (dependent).
- **Multiple Linear Regression**: Two or more independent variables, one dependent variable.
  - **Example**: Size, number of bedrooms, and age (independent variables) predicting house price (dependent).



**Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in
a given dataset?**

**ANSWER:-----**

Linear regression relies on several key assumptions to ensure the validity and reliability of the model's results. Here are the main assumptions and how to check if they hold in a given dataset:

### Assumptions of Linear Regression

1. **Linearity**: The relationship between the independent variables and the dependent variable should be linear.
2. **Independence**: The residuals (errors) should be independent. There should be no correlation between consecutive residuals (no autocorrelation).
3. **Homoscedasticity**: The residuals should have constant variance at every level of the independent variables. In other words, the spread of the residuals should be constant across all levels of the independent variables.
4. **Normality of Residuals**: The residuals should be approximately normally distributed.
5. **No Multicollinearity**: In the case of multiple linear regression, the independent variables should not be too highly correlated with each other.

### Checking the Assumptions

1. **Linearity**:
   - **Scatter plot**: Plot the dependent variable against each independent variable. Look for a linear relationship.
   - **Residuals vs. Fitted Values Plot**: Plot residuals against predicted values. There should be no discernible pattern.

2. **Independence**:
   - **Durbin-Watson Test**: A statistical test that detects the presence of autocorrelation in the residuals from a regression analysis.
   - **Residual Plot**: Plot residuals over time or against an index. Random scatter indicates independence.

3. **Homoscedasticity**:
   - **Residuals vs. Fitted Values Plot**: The spread of residuals should be roughly the same across all levels of fitted values. Look for any funnel shape (which would indicate heteroscedasticity).
   - **Breusch-Pagan Test**: A statistical test for heteroscedasticity.

4. **Normality of Residuals**:
   - **Histogram of Residuals**: The residuals should follow a bell-shaped curve.
   - **Q-Q Plot**: A quantile-quantile plot compares the distribution of residuals to a normal distribution. The points should lie approximately along the diagonal line.
   - **Shapiro-Wilk Test**: A statistical test to assess the normality of the residuals.

5. **No Multicollinearity**:
   - **Variance Inflation Factor (VIF)**: Calculate VIF for each independent variable. A VIF value greater than 10 indicates high multicollinearity.
   - **Correlation Matrix**: Examine the correlation coefficients between independent variables. High correlation coefficients (close to 1 or -1) suggest multicollinearity.



**Q3. How do you interpret the slope and intercept in a linear regression model? Provide an example using
a real-world scenario.**

**ANSWER:-----**


### Interpreting the Slope and Intercept in a Linear Regression Model

In a linear regression model, the equation is typically written as:

\[ y = \beta_0 + \beta_1x + \epsilon \]

Where:
- \( y \) is the dependent variable.
- \( x \) is the independent variable.
- \( \beta_0 \) is the intercept.
- \( \beta_1 \) is the slope.
- \( \epsilon \) is the error term.

#### **Intercept (\( \beta_0 \))**
The intercept is the value of the dependent variable \( y \) when the independent variable \( x \) is zero. It represents the baseline level of the dependent variable in the absence of the independent variable.

#### **Slope (\( \beta_1 \))**
The slope represents the change in the dependent variable \( y \) for a one-unit change in the independent variable \( x \). It indicates the direction and magnitude of the relationship between \( x \) and \( y \).

### Example: Predicting House Prices

Let's consider a real-world scenario where we are predicting the price of a house based on its size (in square feet).

#### Regression Equation
\[ \text{Price} = 50,000 + 150 \times \text{Size} \]

Where:
- **Price**: The predicted price of the house.
- **Size**: The size of the house in square feet.
- **50,000**: The intercept (\( \beta_0 \)).
- **150**: The slope (\( \beta_1 \)).

#### Interpreting the Intercept (\( \beta_0 \))
- The intercept is 50,000. This means that if the size of the house is 0 square feet (hypothetically), the base price of the house would be $50,000. This could be considered the starting price for the house or the value attributed to factors not included in the model (like land value, location, etc.).

#### Interpreting the Slope (\( \beta_1 \))
- The slope is 150. This means that for each additional square foot of house size, the price of the house increases by $150. In other words, if a house's size increases by 1,000 square feet, the price is expected to increase by $150,000 (1000 square feet * $150 per square foot).

### Putting It Together

#### Example Calculation
If we want to predict the price of a house that is 2,000 square feet, we can plug the size into the regression equation:

\[ \text{Price} = 50,000 + 150 \times 2000 \]
\[ \text{Price} = 50,000 + 300,000 \]
\[ \text{Price} = 350,000 \]

Therefore, a house that is 2,000 square feet is predicted to cost $350,000 based on this model.

### Summary
- **Intercept**: Represents the baseline value of the dependent variable when the independent variable is zero.
- **Slope**: Indicates the rate of change in the dependent variable for a one-unit change in the independent variable.

These interpretations help in understanding the influence of the independent variable on the dependent variable and in making predictions based on the regression model.

**Q4. Explain the concept of gradient descent. How is it used in machine learning?**

**ANSWER:------**

### Concept of Gradient Descent

Gradient descent is an optimization algorithm used to minimize the cost (or loss) function in machine learning and statistical modeling. The main idea is to iteratively adjust the model parameters to find the values that minimize the cost function.

#### Key Concepts

1. **Cost Function**: A function that measures the error between the predicted values and the actual values. The goal of gradient descent is to minimize this function.
   
   - In linear regression, the cost function is typically the Mean Squared Error (MSE):
     \[
     J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2
     \]
     Where:
     - \( m \) is the number of training examples.
     - \( h_\theta(x^{(i)}) \) is the predicted value.
     - \( y^{(i)} \) is the actual value.

2. **Parameters**: These are the values that the model uses to make predictions. In linear regression, these parameters are the coefficients (weights) and the intercept.

3. **Learning Rate (\( \alpha \))**: A hyperparameter that determines the size of the steps taken to reach the minimum of the cost function. If the learning rate is too small, the algorithm will take a long time to converge. If it's too large, the algorithm might overshoot the minimum and fail to converge.

#### Gradient Descent Algorithm

1. **Initialize Parameters**: Start with initial guesses for the model parameters.
2. **Compute the Gradient**: Calculate the gradient (partial derivatives) of the cost function with respect to each parameter. The gradient indicates the direction and rate of the steepest increase in the cost function.
   \[
   \frac{\partial J(\theta)}{\partial \theta_j}
   \]
3. **Update Parameters**: Adjust the parameters in the direction opposite to the gradient. This step is repeated until convergence (when the changes in the cost function are below a certain threshold).
   \[
   \theta_j := \theta_j - \alpha \frac{\partial J(\theta)}{\partial \theta_j}
   \]

### Types of Gradient Descent

1. **Batch Gradient Descent**: Uses the entire dataset to compute the gradient and update the parameters.
2. **Stochastic Gradient Descent (SGD)**: Uses a single training example to compute the gradient and update the parameters. It can converge faster but introduces more noise in the updates.
3. **Mini-batch Gradient Descent**: Uses a subset (mini-batch) of the dataset to compute the gradient and update the parameters. It balances the trade-offs between batch gradient descent and SGD.

### Usage in Machine Learning

Gradient descent is widely used in various machine learning algorithms to optimize model parameters. Here are some applications:

1. **Linear Regression**: Minimizes the Mean Squared Error between the predicted and actual values.
2. **Logistic Regression**: Minimizes the log-loss function to find the best decision boundary for classification.
3. **Neural Networks**: Optimizes the weights and biases in each layer by minimizing the cost function (e.g., cross-entropy loss). The backpropagation algorithm uses gradient descent to update the weights.
4. **Support Vector Machines (SVMs)**: Used to optimize the margin between different classes.

### Example: Gradient Descent in Linear Regression

Suppose we have a simple linear regression model:

\[ h_\theta(x) = \theta_0 + \theta_1x \]

The cost function (MSE) is:

\[ J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2 \]

#### Gradient Descent Steps:

1. **Initialize** \( \theta_0 \) and \( \theta_1 \) (e.g., 0).
2. **Compute the gradients**:
   \[
   \frac{\partial J(\theta)}{\partial \theta_0} = \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})
   \]
   \[
   \frac{\partial J(\theta)}{\partial \theta_1} = \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})x^{(i)}
   \]
3. **Update the parameters**:
   \[
   \theta_0 := \theta_0 - \alpha \frac{\partial J(\theta)}{\partial \theta_0}
   \]
   \[
   \theta_1 := \theta_1 - \alpha \frac{\partial J(\theta)}{\partial \theta_1}
   \]
4. **Repeat** until convergence.

### Visualization of Gradient Descent

To better understand gradient descent, visualize it as a ball rolling down a hill. The slope of the hill represents the gradient, and the ball's position represents the current parameter values. The ball moves in the direction that reduces the slope (cost function) the most, eventually reaching the bottom (minimum cost).

### Conclusion

Gradient descent is a fundamental optimization technique in machine learning, enabling models to learn from data by iteratively adjusting parameters to minimize the cost function. Its effectiveness and efficiency make it a cornerstone in training various machine learning models.

**Q5. Describe the multiple linear regression model. How does it differ from simple linear regression?**

**ANSWER:-------**

### Multiple Linear Regression Model

**Multiple linear regression** is an extension of simple linear regression that allows for the modeling of a linear relationship between a dependent variable and two or more independent variables. It is used when the dependent variable is assumed to be affected by more than one predictor.

#### Model Equation

The general form of the multiple linear regression model is:

\[ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \ldots + \beta_nx_n + \epsilon \]

Where:
- \( y \) is the dependent variable.
- \( x_1, x_2, \ldots, x_n \) are the independent variables.
- \( \beta_0 \) is the intercept.
- \( \beta_1, \beta_2, \ldots, \beta_n \) are the coefficients of the independent variables.
- \( \epsilon \) is the error term.

### Differences Between Simple and Multiple Linear Regression

#### 1. Number of Independent Variables
- **Simple Linear Regression**: Involves one independent variable.
  - **Equation**: \( y = \beta_0 + \beta_1x + \epsilon \)
- **Multiple Linear Regression**: Involves two or more independent variables.
  - **Equation**: \( y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \ldots + \beta_nx_n + \epsilon \)

#### 2. Model Complexity
- **Simple Linear Regression**: Simpler and easier to visualize since it deals with only one predictor.
- **Multiple Linear Regression**: More complex, involving multiple predictors, which can capture more intricate relationships in the data.

#### 3. Interpretation of Coefficients
- **Simple Linear Regression**: The slope (\( \beta_1 \)) represents the change in the dependent variable for a one-unit change in the independent variable.
- **Multiple Linear Regression**: Each coefficient (\( \beta_i \)) represents the change in the dependent variable for a one-unit change in the corresponding independent variable (\( x_i \)), holding all other variables constant.

#### 4. Use Cases
- **Simple Linear Regression**: Used when the dependent variable is influenced primarily by one predictor.
  - **Example**: Predicting a student's exam score based on the number of hours studied.
- **Multiple Linear Regression**: Used when the dependent variable is influenced by multiple predictors.
  - **Example**: Predicting house prices based on factors like size, number of bedrooms, age, and location.

### Example of Multiple Linear Regression

Suppose we want to predict the price of a house based on its size (in square feet), number of bedrooms, and age. The model might look like this:

\[ \text{Price} = \beta_0 + \beta_1 \times \text{Size} + \beta_2 \times \text{Bedrooms} + \beta_3 \times \text{Age} + \epsilon \]

#### Interpretation of Coefficients:
- **\(\beta_0\)**: The intercept, representing the baseline price of the house when size, number of bedrooms, and age are all zero (hypothetical baseline).
- **\(\beta_1\)**: The increase in price for each additional square foot of size, holding the number of bedrooms and age constant.
- **\(\beta_2\)**: The increase in price for each additional bedroom, holding size and age constant.
- **\(\beta_3\)**: The decrease in price for each additional year of the house's age, holding size and number of bedrooms constant.

### Summary

- **Simple Linear Regression**: One independent variable, straightforward to interpret and visualize, used when a single predictor is sufficient.
- **Multiple Linear Regression**: Multiple independent variables, more complex, can capture the combined effect of several predictors on the dependent variable.

Multiple linear regression provides a more comprehensive understanding of the factors influencing the dependent variable by considering multiple predictors simultaneously, making it a powerful tool for more accurate and detailed modeling in various fields.

**Q6. Explain the concept of multicollinearity in multiple linear regression. How can you detect and
address this issue?**

**ANSWER:-----**


### Concept of Multicollinearity in Multiple Linear Regression

**Multicollinearity** occurs in multiple linear regression when two or more independent variables are highly correlated with each other. This means that one independent variable can be linearly predicted from the others with a substantial degree of accuracy. Multicollinearity can make it difficult to determine the individual effect of each predictor on the dependent variable, leading to unreliable estimates of the regression coefficients.

#### Problems Caused by Multicollinearity
1. **Inflated Standard Errors**: When multicollinearity is present, the standard errors of the coefficients are inflated, making it more difficult to determine if the predictor variables are statistically significant.
2. **Unstable Coefficients**: The regression coefficients can become very sensitive to small changes in the model, resulting in large variations in the estimated coefficients.
3. **Reduced Interpretability**: High correlations between predictors make it hard to assess the impact of each predictor variable individually.

### Detecting Multicollinearity

1. **Correlation Matrix**:
   - A correlation matrix displays the correlation coefficients between pairs of variables. High correlation coefficients (close to 1 or -1) indicate potential multicollinearity.
   
   
2. **Variance Inflation Factor (VIF)**:
   - VIF quantifies the extent of multicollinearity by measuring how much the variance of a regression coefficient is inflated due to multicollinearity.
   - VIF is calculated as:
     \[
     VIF_j = \frac{1}{1 - R_j^2}
     \]
     Where \( R_j^2 \) is the R-squared value obtained by regressing the \( j \)-th predictor on all the other predictors.
   - A VIF value greater than 10 is often considered indicative of significant multicollinearity.


3. **Eigenvalues and Condition Index**:
   - Multicollinearity can also be detected by examining the eigenvalues of the correlation matrix of the predictors. A condition index (CI) above 30 indicates potential multicollinearity issues.

 
### Addressing Multicollinearity

1. **Remove Highly Correlated Predictors**:
   - Identify and remove one of the highly correlated predictors from the model. This can reduce multicollinearity and improve the stability of the coefficients.

2. **Combine Correlated Predictors**:
   - Combine highly correlated predictors into a single predictor. For example, if two variables are both measuring a similar concept, creating an index or average of the two can be helpful.

3. **Principal Component Analysis (PCA)**:
   - PCA is a dimensionality reduction technique that transforms the correlated predictors into a smaller number of uncorrelated components, which can then be used in the regression model.

 
4. **Ridge Regression**:
   - Ridge regression (a type of regularization) can mitigate the effects of multicollinearity by adding a penalty term to the regression that shrinks the coefficients. This helps to reduce the variance of the coefficient estimates.

### Summary

- **Multicollinearity**: Occurs when independent variables are highly correlated, leading to unreliable and unstable coefficient estimates.
- **Detection Methods**: Correlation matrix, Variance Inflation Factor (VIF), eigenvalues, and condition index.
- **Solutions**: Remove or combine correlated predictors, use PCA for dimensionality reduction, or apply regularization techniques like ridge regression.



**Q7. Describe the polynomial regression model. How is it different from linear regression?**

**ANSWER:-------**


### Polynomial Regression Model

**Polynomial regression** is an extension of linear regression that models the relationship between the independent variable \( x \) and the dependent variable \( y \) as an \( n \)-th degree polynomial. While linear regression assumes a linear relationship between \( x \) and \( y \), polynomial regression allows for a more flexible, curved relationship.

#### Model Equation

The general form of a polynomial regression model is:

\[ y = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3 + \ldots + \beta_nx^n + \epsilon \]

Where:
- \( y \) is the dependent variable.
- \( x \) is the independent variable.
- \( \beta_0, \beta_1, \ldots, \beta_n \) are the coefficients.
- \( \epsilon \) is the error term.
- \( n \) is the degree of the polynomial.

### Differences Between Polynomial Regression and Linear Regression

#### 1. Nature of the Relationship
- **Linear Regression**: Models a straight-line relationship between \( x \) and \( y \).
  - **Equation**: \( y = \beta_0 + \beta_1x + \epsilon \)
- **Polynomial Regression**: Models a curved relationship, which can be more complex and capture non-linear patterns in the data.
  - **Equation**: \( y = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3 + \ldots + \beta_nx^n + \epsilon \)

#### 2. Flexibility
- **Linear Regression**: Limited to linear relationships and might not perform well if the true relationship is non-linear.
- **Polynomial Regression**: More flexible and capable of fitting a wide range of non-linear patterns.

#### 3. Overfitting
- **Linear Regression**: Less prone to overfitting since it models a simple linear relationship.
- **Polynomial Regression**: Higher-degree polynomials can lead to overfitting, where the model captures noise in the data rather than the underlying pattern.

#### 4. Interpretability
- **Linear Regression**: Coefficients are straightforward to interpret.
- **Polynomial Regression**: Coefficients become less interpretable as the degree of the polynomial increases.

### Example of Polynomial Regression

Suppose we have data showing the relationship between the number of hours studied and exam scores, and we suspect that the relationship is non-linear.

#### Data
\[
\begin{array}{|c|c|}
\hline
\text{Hours Studied (x)} & \text{Exam Score (y)} \\
\hline
1 & 50 \\
2 & 55 \\
3 & 65 \\
4 & 70 \\
5 & 75 \\
6 & 78 \\
7 & 80 \\
8 & 85 \\
\hline
\end{array}
\]

#### Polynomial Regression Model
We fit a polynomial regression model of degree 2 (quadratic) to the data:

\[ y = \beta_0 + \beta_1x + \beta_2x^2 + \epsilon \]

After fitting the model, we get the coefficients:

\[ \text{Exam Score} = 40 + 5x + 0.5x^2 \]

### Comparison with Linear Regression

#### Linear Regression Model
\[ y = \beta_0 + \beta_1x + \epsilon \]

If we fit a linear regression model to the same data, we might get:

\[ \text{Exam Score} = 45 + 5x \]

#### Visual Comparison

- **Linear Regression**: The fitted line might not capture the curvature in the data, leading to underfitting.
- **Polynomial Regression**: The fitted curve can better capture the non-linear pattern, providing a more accurate fit.


### Summary

- **Polynomial Regression**: Extends linear regression to model non-linear relationships using polynomial terms of the independent variable.
- **Differences**: Polynomial regression is more flexible, can fit more complex patterns, but is also more prone to overfitting compared to linear regression.
- **Use Case**: When the relationship between the variables is non-linear, polynomial regression can provide a better fit than linear regression.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Example data
x = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
y = np.array([50, 55, 65, 70, 75, 78, 80, 85])

# Polynomial regression (degree 2)
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

# Linear regression
model_linear = LinearRegression()
model_linear.fit(x, y)
y_linear_pred = model_linear.predict(x)

# Plotting
plt.scatter(x, y, color='blue')
plt.plot(x, y_poly_pred, color='red', label='Polynomial Regression (degree 2)')
plt.plot(x, y_linear_pred, color='green', label='Linear Regression')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.legend()
plt.show()
```

**Q8. What are the advantages and disadvantages of polynomial regression compared to linear
regression? In what situations would you prefer to use polynomial regression?**

**ANSWER:------**


### Advantages and Disadvantages of Polynomial Regression Compared to Linear Regression

#### Advantages of Polynomial Regression

1. **Flexibility**:
   - Polynomial regression can model non-linear relationships between the independent and dependent variables, providing a better fit for datasets where the relationship is not strictly linear.

2. **Better Fit for Curved Data**:
   - It can capture the curvature in data, leading to a more accurate representation of the underlying trends, especially in cases where the relationship between variables is inherently non-linear.

3. **Higher Accuracy**:
   - By including higher-degree terms, polynomial regression can reduce the error in prediction, potentially leading to higher R-squared values and lower residual errors.

#### Disadvantages of Polynomial Regression

1. **Overfitting**:
   - Polynomial regression, especially with high-degree polynomials, can overfit the data. It can capture noise and outliers, resulting in poor generalization to new data.
   - Overfitting is more likely when the polynomial degree is too high relative to the amount of data.

2. **Interpretability**:
   - The coefficients of higher-degree polynomial terms are difficult to interpret, making the model less transparent and harder to understand compared to linear regression.

3. **Computational Complexity**:
   - Polynomial regression can become computationally expensive as the degree of the polynomial increases, especially with large datasets.

4. **Multicollinearity**:
   - Higher-degree polynomial terms are often highly correlated, leading to multicollinearity issues that can make the estimation of coefficients unstable and inflate their standard errors.

### Situations to Prefer Polynomial Regression

1. **Non-linear Relationships**:
   - When the relationship between the independent and dependent variables is non-linear, polynomial regression is preferable as it can capture the curvature and provide a better fit.

2. **Data Exploration and Trend Analysis**:
   - When exploring data and looking for trends, polynomial regression can help identify and model the underlying patterns that might not be apparent with linear regression.

3. **Predictive Accuracy for Curved Data**:
   - When the primary goal is to achieve higher predictive accuracy in scenarios where the data exhibits a curved pattern, polynomial regression can be more effective than linear regression.

### Example: Comparing Linear and Polynomial Regression

#### Scenario: Predicting House Prices

Suppose we have a dataset with house prices based on the size of the house. The relationship between house size and price is non-linear, with larger houses seeing diminishing returns in price.

#### Linear Regression
- **Model**: \( \text{Price} = \beta_0 + \beta_1 \times \text{Size} \)
- **Fit**: The model might not capture the non-linear relationship, leading to higher errors.

#### Polynomial Regression (Degree 2)
- **Model**: \( \text{Price} = \beta_0 + \beta_1 \times \text{Size} + \beta_2 \times \text{Size}^2 \)
- **Fit**: The model can capture the diminishing returns, providing a better fit and more accurate predictions.



### Summary

- **Advantages of Polynomial Regression**: Flexibility, better fit for curved data, and higher accuracy for non-linear relationships.
- **Disadvantages**: Prone to overfitting, harder to interpret, computationally complex, and potential multicollinearity issues.
- **Preferred Situations**: Non-linear relationships, data exploration, and trend analysis where the data exhibits curvature.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Example data
size = np.array([500, 1000, 1500, 2000, 2500, 3000, 3500, 4000]).reshape(-1, 1)
price = np.array([150000, 200000, 270000, 320000, 360000, 390000, 410000, 420000])

# Linear regression
linear_model = LinearRegression()
linear_model.fit(size, price)
price_linear_pred = linear_model.predict(size)

# Polynomial regression (degree 2)
poly = PolynomialFeatures(degree=2)
size_poly = poly.fit_transform(size)
poly_model = LinearRegression()
poly_model.fit(size_poly, price)
price_poly_pred = poly_model.predict(size_poly)

# Plotting
plt.scatter(size, price, color='blue')
plt.plot(size, price_linear_pred, color='green', label='Linear Regression')
plt.plot(size, price_poly_pred, color='red', label='Polynomial Regression (degree 2)')
plt.xlabel('Size of the House (sq ft)')
plt.ylabel('Price of the House')
plt.legend()
plt.show()
```
